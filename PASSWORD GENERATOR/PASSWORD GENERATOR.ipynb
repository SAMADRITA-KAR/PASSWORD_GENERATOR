```python
# Generating a random password
import secrets
import string
L=string.ascii_letters
D=string.digits
SC=string.punctuation
CH=L+D+SC
p_len=int(input("How many characters should your password have? "))
p=""
for i in range(p_len):
  p+= "".join(secrets.choice(CH))
print("Your password is ",p)

# (Doubt - Exactly where and how to add constraints to this code snippet??)

# Splitting every character of the password using Unpack(*) function
a=[*p]
print([*p])

for i in range(0,p_len+1):
  a.replace('A','Z')
  a.replace('B','Y')
```
